```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('vgsales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(figsize=(20,15), grid=False, layout=(2,4), bins=10)
plt.show()
```

```python
plt.figure(figsize=(25,10))
sns.countplot(df['Year'])
plt.show()
```

```python
df.boxplot(figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(df['Global_Sales'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
figure = px.scatter(df,
                x='NA_Sales',
                y='Year',
                hover_name='Name',
                title='NA_Sales vs. Year')
figure.show()
```
